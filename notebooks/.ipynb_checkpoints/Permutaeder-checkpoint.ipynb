```python
P = polytopes.permutahedron(4)
Q = P.affine_hull_projection(orthonormal=True, extend=True)
n = len(Q.faces(2))
n
```

```python
P.show()
```

```python
n = 6 + 8
my_rainbow = rainbow(n)
shuffle(my_rainbow)
```

```python
separated_faces = dict()
vertex_coloring = Q.graph().coloring()
centers = []
c = 0
for face in Q.faces(2):
    if len(face.vertices()) == 4:
        separated_faces[(face.as_polyhedron())] = my_rainbow[c]
        c += 1
    elif len(face.vertices()) == 6:
        center = sum(map(vector, face.vertices()))/6
        centers.append(center)
        for i, vertex in enumerate(face.as_polyhedron().faces(0)):
            #print(vertex)
            if vertex.vertices()[0] in vertex_coloring[0]:
                two_edges = [edge for edge in face.as_polyhedron().faces(1) if edge.as_polyhedron().contains(vertex.vertices()[0])]
                assert len(two_edges) == 2
                hex = Polyhedron([center] + [vertex for edge in two_edges for vertex in edge.vertices()])
                separated_faces[hex] = my_rainbow[c]
        c += 1
                
    else:
        print("should not happen")
```

```python
other_generators = dict()

for vertex in vertex_coloring[0] + centers:
    other_generators[tuple(list(vertex))] = tuple(face for face in separated_faces.keys() if face.contains(vertex))
```

```python
all_faces = dict((face, i) for i, face in enumerate(separated_faces.keys()))
other_gens  = list(map(lambda faces : tuple(map(lambda i : all_faces[i] + 1, faces)), other_generators.values()))
other_gens
```

```python
Perms = PermutationGroup(other_gens)
Perms.order()
```

```python
point_strings = []
name_to_perm = dict()

for i, (vertex, faces) in enumerate(other_generators.items()):
    vertex_index = i
    print(vertex, vertex_index)
    name_to_perm[vertex_index] = tuple(map(lambda i : all_faces[i], faces))
    point_strings.append(f'{{"point": {list(map(float, list(vertex)))}, "size":20.0, "color": "#000000", "opacity": 10.0, name:"{vertex_index}"}}')
```
